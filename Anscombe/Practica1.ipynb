```python
import pandas as pd

# Cargar el archivo CSV (aseg√∫rate de que el nombre del archivo es correcto)
df = pd.read_csv('/content/df_anscombe.csv')

# Visualizar las primeras filas del dataset
print(df.head())
```

```python
df.groupby('group').describe()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un gr√°fico para cada categor√≠a en 'group'
groups = df['group'].unique()

# Definir el tama√±o del gr√°fico
plt.figure(figsize=(10, 6))

# Graficar cada grupo en un subplot diferente
for i, group in enumerate(groups, 1):
    plt.subplot(2, 2, i)  # 2x2 subplot (cambiar seg√∫n el n√∫mero de categor√≠as)
    subset = df[df['group'] == group]
    sns.scatterplot(x='x', y='y', data=subset)
    plt.title(f'Grupo {group}')
    plt.xlabel('x')
    plt.ylabel('y')

# Ajustar el layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Definir el tama√±o del gr√°fico
plt.figure(figsize=(10, 6))

# Crear un gr√°fico de cajas y bigotes para cada grupo
sns.boxplot(x='group', y='y', data=df)

# Agregar t√≠tulos y etiquetas
plt.title('Boxplot por categor√≠a')
plt.xlabel('Grupo')
plt.ylabel('Valores de y')

# Mostrar el gr√°fico
plt.show()
```
